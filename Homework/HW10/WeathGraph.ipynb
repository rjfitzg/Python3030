```python
import pandas as pd
import seaborn as sb # seaborn used matplotlib as a backend

df_all = pd.read_csv('weath_mod.csv')
```

```python
for i in range(1,12+1):
    x = df_all[df_all['Month'] == i]
    sb.lineplot(x['Day'], df_all['Temperature daily mean [2 m above gnd]'])
```
